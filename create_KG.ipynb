```python
# Function to convert CSV data to RDF using the schema
def create_dance_knowledge_graph(dataframe):
    """
    Convert dance CSV data to RDF knowledge graph
    """
    kg = Graph()

    # Bind namespaces
    kg.bind("schema", SCHEMA)
    kg.bind("dance", DANCE)
    kg.bind("dbp", DBP)
    kg.bind("dbo", DBO)

    for idx, row in dataframe.iterrows():
        # Create unique URI for each dance style
        dance_name = row.get('Dance style', f'dance_{idx}')
        dance_uri = DANCE[quote(dance_name.replace(' ', '_'))]

        # Add type
        kg.add((dance_uri, RDF.type, DANCE.DanceStyle))

        # Add properties
        if pd.notna(row.get('Dance style')):
            kg.add((dance_uri, SCHEMA.name, Literal(row['Dance style'], datatype=XSD.string)))

        if pd.notna(row.get('Dance Type')):
            kg.add((dance_uri, DANCE.danceType, Literal(row['Dance Type'], datatype=XSD.string)))

        if pd.notna(row.get('Origin')):
            # Create Place entity
            place_uri = SCHEMA[quote(row['Origin'].replace(' ', '_'))]
            kg.add((place_uri, RDF.type, SCHEMA.Place))
            kg.add((place_uri, SCHEMA.name, Literal(row['Origin'], datatype=XSD.string)))
            kg.add((dance_uri, SCHEMA.spatialCoverage, place_uri))

        if pd.notna(row.get('Time Period')):
            # Create TimePeriod entity
            period_uri = DANCE[quote(row['Time Period'].replace(' ', '_'))]
            kg.add((period_uri, RDF.type, DANCE.TimePeriod))
            kg.add((period_uri, DANCE.periodName, Literal(row['Time Period'], datatype=XSD.string)))
            kg.add((dance_uri, SCHEMA.temporalCoverage, period_uri))

        if pd.notna(row.get('Cultural Significance')):
            kg.add((dance_uri, DANCE.culturalSignificance, Literal(row['Cultural Significance'], datatype=XSD.string)))

        if pd.notna(row.get('Notable Characteristics')):
            kg.add((dance_uri, DANCE.characteristics, Literal(row['Notable Characteristics'], datatype=XSD.string)))

        if pd.notna(row.get('Instrumental')):
            # Create Instrument entities
            instruments = str(row['Instrumental']).split(',')
            for instrument in instruments:
                instrument = instrument.strip()
                if instrument:
                    instrument_uri = SCHEMA[quote(instrument.replace(' ', '_'))]
                    kg.add((instrument_uri, RDF.type, SCHEMA.MusicInstrument))
                    kg.add((instrument_uri, DANCE.instrumentName, Literal(instrument, datatype=XSD.string)))
                    kg.add((dance_uri, DANCE.hasInstrument, instrument_uri))

        if pd.notna(row.get('Hardness Ratio')):
            try:
                kg.add((dance_uri, DANCE.difficultyLevel, Literal(float(row['Hardness Ratio']), datatype=XSD.float)))
            except:
                pass

        if pd.notna(row.get('Dance Formation')):
            kg.add((dance_uri, DANCE.formation, Literal(row['Dance Formation'], datatype=XSD.string)))

        if pd.notna(row.get('Costume')):
            kg.add((dance_uri, DANCE.costume, Literal(row['Costume'], datatype=XSD.string)))

        if pd.notna(row.get('Tempo (BPM)')):
            try:
                kg.add((dance_uri, DANCE.tempo, Literal(int(row['Tempo (BPM)']), datatype=XSD.integer)))
            except:
                pass

        if pd.notna(row.get('Famous Practitioners')):
            # Create Person entities
            practitioners = str(row['Famous Practitioners']).split(',')
            for practitioner in practitioners:
                practitioner = practitioner.strip()
                if practitioner:
                    person_uri = SCHEMA[quote(practitioner.replace(' ', '_'))]
                    kg.add((person_uri, RDF.type, SCHEMA.Person))
                    kg.add((person_uri, SCHEMA.name, Literal(practitioner, datatype=XSD.string)))
                    kg.add((dance_uri, SCHEMA.performer, person_uri))

        if pd.notna(row.get('Events and Festivals')):
            # Create Event entities (detect if it's a competition or festival)
            events = str(row['Events and Festivals']).split(',')
            for event in events:
                event = event.strip()
                if event:
                    event_uri = SCHEMA[quote(event.replace(' ', '_'))]
                    # Determine type based on keywords
                    event_lower = event.lower()
                    if 'competition' in event_lower or 'contest' in event_lower or 'championship' in event_lower:
                        kg.add((event_uri, RDF.type, DANCE.Competition))
                    elif 'festival' in event_lower or 'celebration' in event_lower:
                        kg.add((event_uri, RDF.type, DANCE.Festival))
                    else:
                        kg.add((event_uri, RDF.type, SCHEMA.Event))
                    kg.add((event_uri, DANCE.eventName, Literal(event, datatype=XSD.string)))
                    kg.add((dance_uri, SCHEMA.event, event_uri))

        if pd.notna(row.get('Modern Adaptations')):
            kg.add((dance_uri, DANCE.modernAdaptations, Literal(row['Modern Adaptations'], datatype=XSD.string)))

        if pd.notna(row.get('Associated Music Genre')):
            # Create MusicGenre entity
            genres = str(row['Associated Music Genre']).split(',')
            for genre in genres:
                genre = genre.strip()
                if genre:
                    genre_uri = SCHEMA[quote(genre.replace(' ', '_'))]
                    kg.add((genre_uri, RDF.type, SCHEMA.MusicGenre))
                    kg.add((genre_uri, DANCE.genreName, Literal(genre, datatype=XSD.string)))
                    kg.add((dance_uri, SCHEMA.genre, genre_uri))

        if pd.notna(row.get('Learning Difficulty')):
            # Create DifficultyLevel entity
            difficulty = str(row['Learning Difficulty']).strip()
            difficulty_uri = DANCE[quote(difficulty.replace(' ', '_'))]
            kg.add((difficulty_uri, RDF.type, DANCE.DifficultyLevel))
            kg.add((difficulty_uri, DANCE.difficultyName, Literal(difficulty, datatype=XSD.string)))
            kg.add((dance_uri, DANCE.hasDifficultyLevel, difficulty_uri))

        if pd.notna(row.get('Health Benefits')):
            # Create HealthBenefit entities (parse comma-separated benefits)
            benefits = str(row['Health Benefits']).split(',')
            for benefit in benefits:
                benefit = benefit.strip()
                if benefit:
                    benefit_uri = DANCE[quote(benefit.replace(' ', '_'))]
                    kg.add((benefit_uri, RDF.type, DANCE.HealthBenefit))
                    kg.add((benefit_uri, DANCE.benefitName, Literal(benefit, datatype=XSD.string)))

                    # Auto-detect category based on keywords
                    benefit_lower = benefit.lower()
                    if any(word in benefit_lower for word in ['cardiovascular', 'fitness', 'flexibility', 'coordination', 'posture', 'balance', 'physical']):
                        kg.add((benefit_uri, DANCE.benefitCategory, Literal("Physical", datatype=XSD.string)))
                    elif any(word in benefit_lower for word in ['stress', 'mental', 'cognitive', 'confidence', 'self-esteem', 'mood']):
                        kg.add((benefit_uri, DANCE.benefitCategory, Literal("Mental", datatype=XSD.string)))
                    elif any(word in benefit_lower for word in ['social', 'community', 'connection']):
                        kg.add((benefit_uri, DANCE.benefitCategory, Literal("Social", datatype=XSD.string)))
                    else:
                        kg.add((benefit_uri, DANCE.benefitCategory, Literal("General", datatype=XSD.string)))

                    kg.add((dance_uri, DANCE.hasHealthBenefit, benefit_uri))

        if pd.notna(row.get('Age Group')):
            # Create AgeGroup entity
            age_groups = str(row['Age Group']).split(',')
            for age_group in age_groups:
                age_group = age_group.strip()
                if age_group:
                    age_group_uri = DANCE[quote(age_group.replace(' ', '_'))]
                    kg.add((age_group_uri, RDF.type, DANCE.AgeGroup))
                    kg.add((age_group_uri, DANCE.ageGroupName, Literal(age_group, datatype=XSD.string)))
                    kg.add((dance_uri, DANCE.hasAgeGroup, age_group_uri))

    return kg

# Create the knowledge graph from CSV data
print("\n" + "=" * 80)
print("CREATING KNOWLEDGE GRAPH FROM DATA")
print("=" * 80)

dance_kg = create_dance_knowledge_graph(dat_Dancing)
print(f"✓ Knowledge graph created with {len(dance_kg)} triples")

# Save the knowledge graph
kg_output_path = '/Users/hannes/Documents/GitHub/KG_Personal_Project/dance_knowledge_graph.ttl'
dance_kg.serialize(destination=kg_output_path, format='turtle')
print(f"✓ Knowledge graph saved to: {kg_output_path}")

# Print a sample of the graph
print("\n" + "=" * 80)
print("SAMPLE OF KNOWLEDGE GRAPH (first 50 triples)")
print("=" * 80)
sample_triples = list(dance_kg)[:50]
for s, p, o in sample_triples:
    print(f"{s} -> {p} -> {o}")
```
