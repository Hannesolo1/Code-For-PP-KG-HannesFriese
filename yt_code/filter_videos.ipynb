```python
import pandas as pd
from pathlib import Path

CSV_PATH = Path("../data/yt_data/dance_videos.csv")

df = pd.read_csv(CSV_PATH)
print(f"Rows before filtering: {len(df)}")
print(f"Unique (dance_style, dance_type) combos: {df[['dance_style','dance_type']].drop_duplicates().shape[0]}")
df.head(3)

# ── Ensure numeric types ───────────────────────────────────────────────────────
df['num_comments_retrieved']       = pd.to_numeric(df['num_comments_retrieved'],       errors='coerce').fillna(0)
df['comment_sentiment_avg_compound'] = pd.to_numeric(df['comment_sentiment_avg_compound'], errors='coerce').fillna(0)

# ── Keep exactly one video per (dance_style, dance_type) ───────────────────────
# Priority 1: highest num_comments_retrieved
# Priority 2: highest comment_sentiment_avg_compound (tiebreaker)

filtered = (
    df
    .sort_values(
        by=['dance_style', 'dance_type', 'num_comments_retrieved', 'comment_sentiment_avg_compound'],
        ascending=[True, True, False, False]
    )
    .drop_duplicates(subset=['dance_style', 'dance_type'], keep='first')
    .reset_index(drop=True)
)

print(f"Rows after filtering : {len(filtered)}")
print(f"Unique combos kept   : {filtered[['dance_style','dance_type']].drop_duplicates().shape[0]}")
filtered[['dance_style','dance_type','video_id','num_comments_retrieved','comment_sentiment_avg_compound']].head(10)

# ── Save result ────────────────────────────────────────────────────────────────
OUT_PATH = Path("../data/yt_data/dance_videos_filtered.csv")
filtered.to_csv(OUT_PATH, index=False, encoding='utf-8')
print(f"Saved to: {OUT_PATH}")
```
